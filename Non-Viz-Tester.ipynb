```python
import numpy
import mesa
from model import HumanitarianLogistics, COA, AZC
import matplotlib.pyplot as plt
```

```python
canvas_width = 600
canvas_height = 400
num_azc = 3
num_cities = 2
grid_width = int(canvas_width / 4)
grid_height = int(canvas_height / 4)
Test = HumanitarianLogistics(100, 200, .05, 3, 3, .9, grid_width, grid_height)
```

```python
for i in range(50):
    Test.step()
```

```python
coa = [agent for agent in Test.schedule.agents if
      type(agent) is COA and not agent.ta]
```

```python
names = ['Consolidate', 'Invest', 'Segregate', 'Integrate']
hist = {name:0 for name in names}


for c in coa:
    for name in names:
        for act in c.actions:
            if act.name == name:
                hist[name] += act.counter
plt.bar(hist.keys(), hist.values())
plt.show()
```
